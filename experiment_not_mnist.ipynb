```python
from digitnotdigit import DigitNotDigit
from utilities import load_notmnist
import numpy as np
import pandas as pd
```

```python
classifier = DigitNotDigit()
notmnist_dataset, notmnist_label = load_notmnist(normalized=True)
```

```python
predicted_label, digit_label = classifier.predict(notmnist_dataset)
```

```python
# Get all the index with the label 1 (misclassified as digit)
# There are 3430 wrong labels :(
predicted_label.shape, digit_label.shape
wrong_label_idx = np.where(predicted_label == 1)
len(wrong_label_idx[0])

# Let's build "confusion matrix" for this wrong classified data
# notmnist_label (the character A- J) as the column and digit_label (0-9) as the row for this wrong labels
# Init the matrix
conf = np.zeros((10, 10), dtype=np.int)
for wrong_id in wrong_label_idx[0]:
    char_label = notmnist_label[wrong_id]
    d_label = digit_label[wrong_id]
    conf[d_label][char_label] += 1
    
# Show pretty matrix :)
conf_df = pd.DataFrame(conf, columns=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J' ])
conf_df
```
